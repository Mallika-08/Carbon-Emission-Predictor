## ANN

```python
import keras as k
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
trainx = pd.DataFrame(sc.fit_transform(trainx))
testx =pd.DataFrame(sc.transform(testx))
```

```python
trainx.shape
```

```python
trainx[:5]
```

```python
ann=k.models.Sequential()
```

```python
trainx.shape
```

```python
testx.shape
```

```python
ann.add(k.layers.Dense(units=12, activation='relu',input_shape=(None,9)))
ann.add(k.layers.Dense(units=8, activation='relu'))
ann.add(k.layers.Dense(units=1))
```

```python
from tensorflow.keras.utils import plot_model
plot_model(ann,
           to_file="model.png",
           show_shapes=True,
           show_layer_names=True,
          )
```

```python
ann.compile(optimizer='adam',loss = 'mean_squared_error', metrics = ['mse'])
```

```python
history=ann.fit(trainx,trainy,epochs=150,validation_split=0.2, verbose=1,batch_size=32)
```
